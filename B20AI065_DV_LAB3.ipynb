```python
df = pd.read_csv("/content/drive/MyDrive/DV LAB/lab3/Flightschedule.csv")
df=df.drop(0)
df=df.drop(1)
df=df.reset_index()
df.columns=['index','origin','destination','flightnumber','departure','arrival','flighttime','routing','days','connectionflight','effectivefrom','effectiveto']
df
```

```python
df["origin"].unique()
```

```python
df = df[df["origin"]!='Abu Dhabi']
df = df[df["origin"]!='Bangkok']
df = df[df["origin"]!='Dubai']
df = df[df["origin"]!='Kuwait']
df = df[df["origin"]!='Male']
df = df[df["origin"]!='Muscat']
df = df[df["origin"]!='Phuket']
df = df[df["origin"]!='Dammam']
```

```python
df['origin'].unique()
```

```python
df['destination'].unique()
```

```python
df = df[df["destination"]!='Abu Dhabi']
df = df[df["destination"]!='Bangkok']
df = df[df["destination"]!='Dubai']
df = df[df["destination"]!='Kuwait']
df = df[df["destination"]!='Male']
df = df[df["destination"]!='Muscat']
df = df[df["destination"]!='Phuket']
df = df[df["destination"]!='Dammam']
```

```python
df['destination'].unique()
```

```python
df=df.reset_index()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises")
gdf.plot(figsize=(20,20), facecolor='white', edgecolor='green')
routes=[]

for i in range(len(df)):

    if [df["origin"][i],df["destination"][i]] not in routes and [df["destination"][i],df["origin"][i]] not in routes:
      routes.append([df["origin"][i],df["destination"][i]])
      originlat, originlon = geolocator.geocode([df["origin"][i]]).latitude, geolocator.geocode([df["origin"][i]]).longitude
      destlat, destlon = geolocator.geocode([df["destination"][i]]).latitude, geolocator.geocode([df["destination"][i]]).longitude
      plt.plot([originlon, destlon], [originlat, destlat], linewidth=1, color="lime", marker='o')
      plt.text(originlon, originlat, df["origin"][i], color="black", fontsize=14, horizontalalignment="center") 
      plt.text(destlon, destlat, df["destination"][i], color="black", fontsize=14, horizontalalignment="center") 

plt.show()
```
